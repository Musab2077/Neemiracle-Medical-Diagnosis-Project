```python
from sklearn.preprocessing import LabelEncoder
disease_le=LabelEncoder()
prescription_le=LabelEncoder()

disease=disease_le.fit_transform(y['Disease'])
prescription=prescription_le.fit_transform(y['Prescription'])
```

```python
disease_categorical=to_categorical(disease)
prescription_categorical=to_categorical(prescription)
Y=np.hstack((disease_categorical,prescription_categorical))
```

```python
from tensorflow.keras import metrics
from tensorflow.keras.losses import CategoricalCrossentropy
from tensorflow.keras.optimizers import Adadelta
from tensorflow.keras.regularizers import l1
```

```python
# input=Input(shape=(9,))
def make_embedding():
    inp1=Input((12,),name='Input Disease')
    # inp2=Input((9,),name='Input Prescription')
    
    e1=Embedding(input_dim=5000,output_dim=64)(inp1)
    # e2=Embedding(input_dim=5000,output_dim=150,input_shape=(9,))(inp1)
    # d1=Dense(64,activation='sigmoid')(embedding_layer)
    
    lstm1=LSTM(64,activation='leaky_relu')(e1)
    
    # f1=Flatten()(l1)
    
    # concatenate=Concatenate()([l1,l2])
    d1=Dense(512,activation='leaky_relu')(lstm1)
    d2=Dense(512,activation='leaky_relu')(d1)
    # l2=LSTM(32,activation='relu')(e2)
    # d3=Dense(256,activation='relu')(d2)
    # d4=Dense(512,activation='relu')(d3)
    
    disease_output=Dense(len(disease_le.classes_),activation='sigmoid',name='disease_output')(d2)#(concatenate)
    prescription_output=Dense(len(prescription_le.classes_),activation='sigmoid',name='prescription_output')(d2)#(concatenate)
    return Model(inputs=inp1,outputs=[disease_output,prescription_output],name='Embedding')
```

```python
embedding=make_embedding()
embedding.summary()
```

```python
embedding.compile(optimizer='rmsprop',loss={'disease_output':'categorical_crossentropy','prescription_output':'categorical_crossentropy'},
                  metrics={'disease_output': ['accuracy'], 'prescription_output': ['accuracy']})

embedding.fit(x_pad_sequences_scaled,{'disease_output':disease_categorical,
                                      'prescription_output':prescription_categorical},epochs=100,batch_size=50)
```

```python
from tensorflow import expand_dims
example_array=x_pad_sequences_scaled[0]
reshaped_input=expand_dims(example_array,axis=0)
output=embedding(reshaped_input)
```

```python
def for_predicting_outcome(statement):
    tokenized_statement=tokenizer.texts_to_sequences([statement])
    sequenced_statement=pad_sequences(tokenized_statement,12,padding='post')
    
    predicted_outcome=embedding.predict(sequenced_statement)
    
    disease_outcome=np.argmax(predicted_outcome[0],axis=1)[0]
    prescription_outcome=np.argmax(predicted_outcome[1],axis=1)[0]
    
    disease_prediction=disease_le.inverse_transform([disease_outcome])[0]
    prescription_prediction=prescription_le.inverse_transform([prescription_outcome])[0]

    return {'Predicted Disease': disease_prediction,
           'Predicted Prescription': prescription_prediction}
```
